### Problem2:

```python
cancer=load_breast_cancer()
```

```python
 def sample_data(n_used=25):
    # Randomize the order to help get better estimates (there's some structure in the data)
    N=len(cancer.target)
    perm=np.random.permutation(N)
    # limiting data size to emphasize the point
    X=cancer.data[perm,:]
    y=cancer.target[perm]
#     X_holdout=cancer.data[n_used:,:]
#     y_holdout=cancer.target[n_used:]
    X=X[:n_used,:]
    y=y[:n_used]
    return X,y
```

```python
from sklearn import ensemble,model_selection,svm,decomposition,pipeline
cv=model_selection.StratifiedKFold(n_splits=4,shuffle=True,random_state=42)
sample=sample_data()
pca=decomposition.PCA()
clf=svm.SVC(kernel='rbf')
pipe=pipeline.Pipeline([('pca',pca),('clf',clf)])
#raw # of parameters :10*10*10
param_grid={'pca__n_components':np.arange(2,7),'clf__C':np.logspace(-2,0,5),'clf__gamma':np.logspace(-2,0,5)}
gs=model_selection.GridSearchCV(pipe,param_grid,scoring='roc_auc',cv=cv)

perm_mean=[]
perm_std=[]
#from tqdm import trange
np.random.seed(24)
N = [35,50,100,200,400]
for n in N:
    perm_score=[]
    for repeat in range(0,10):
        X,y=sample_data(n)
        y_perm = np.random.permutation(y)
        gs.fit(X,y_perm)
        perm_score.append(gs.best_score_)

    perm_mean.append(np.mean(perm_score))
    perm_std.append(np.std(perm_score))
```

```python
plt.errorbar(N, perm_mean, perm_std, linestyle='None', marker = 'o')
plt.xlabel('N')
plt.ylabel('AUC Score')
plt.title("XXXX")
plt.axis('tight')
plt.show()
```

##### 2(b): When applying more data, AUC of permutation testing is approaching 0.5, and standard deviation is smaller, which means the bias is diminishing.

##### 2(c): I tried 5*5*5= 625 models. Try Increasing search space:

### Problem 3 

```python
co2_df=pd.read_csv('homeworks/hw4/data_co2_weekly_mlo.txt',header=48,delim_whitespace=True)
co2_df[:5]
# some quick_data_cleaning
# replace missing numbers with NaNs for clarity
missing=co2_df['ppm']<0
co2_df['ppm'].values[missing]=np.NaN
times=co2_df['decimal']
co2=co2_df['ppm']
# We have some missing data, so we'll do a quick interpolation.  We'll come back to this.
from scipy.interpolate import interp1d
not_observed=np.isnan(co2)
interpolator=interp1d(times[~not_observed],co2[~not_observed],kind='cubic')
co2_interpolated=interpolator(times)
```

```python
TSS=model_selection.TimeSeriesSplit(n_splits=10)
lin_reg=linear_model.LinearRegression()
```

```python
score_mean_1 = []
score_std_1 = []
for n_lag in np.arange(1,100):
    n_max=len(co2_interpolated)
    X=np.zeros([n_max-n_lag,n_lag])
    for i in range(0,n_max-n_lag):
        X[i]=co2_interpolated[i:i+n_lag]
    y=co2_interpolated[n_lag:]
    scores = []
    for train_ndx,valid_ndx in TSS.split(X,y):
        valid_ndx_adjusted=valid_ndx
        lin_reg.fit(X[train_ndx],y[train_ndx])
        scores.append(metrics.mean_squared_error(y[valid_ndx_adjusted],lin_reg.predict(X[valid_ndx_adjusted])))
    score_mean_1.append(np.mean(scores))
    score_std_1.append(np.std(scores)/np.sqrt(TSS.n_splits))
```

```python
plt.plot(np.arange(1,100),score_mean_1)
plt.xlabel('n_lags')
plt.ylabel('MSE Score')
plt.title("")
plt.axis('tight')
plt.show()
```

```python
print(f'The optimal number of lags is {np.argmin(score_mean_1)}')
```

###### 3(b): Set n_lags to be the same with the optimal number in linear regession in order to compare KNN and Linear Regression Model.

```python
from sklearn import neighbors
min_neighbor = 2
max_neighbor = 30
n_lag = np.argmin(score_mean_1)
score_mean_2 = np.zeros([max_neighbor-min_neighbor+1,n_lag_max-n_lag_min+1])
score_std_2 = np.zeros([max_neighbor-min_neighbor+1,n_lag_max-n_lag_min+1])

n_max=len(co2_interpolated)
X=np.zeros([n_max-n_lag,n_lag])
for i in range(0,n_max-n_lag):
    X[i]=co2_interpolated[i:i+n_lag]
y=co2_interpolated[n_lag:]
knn_score = []
scores = []
for N_neighbor in np.arange(min_neighbor,max_neighbor+1):
    knn=neighbors.KNeighborsRegressor(n_neighbors=N_neighbor)
    for train_ndx,valid_ndx in TSS.split(X,y):
        valid_ndx_adjusted=valid_ndx
        knn.fit(X[train_ndx],y[train_ndx])
        scores.append(metrics.mean_squared_error(y[valid_ndx_adjusted],knn.predict(X[valid_ndx_adjusted])))
    score_mean_2[N_neighbor-min_neighbor,n_lag-n_lag_min] = np.mean(scores)
    score_std_2[N_neighbor-min_neighbor,n_lag-n_lag_min] = np.std(scores)/np.sqrt(TSS.n_splits)
```

```python
plt.plot(np.arange(min_neighbor,max_neighbor+1),score_mean_2.reshape(-1,1)) 
plt.xlabel('n_lags')
plt.ylabel('MSE Score')
plt.title("")
plt.axis('tight')
plt.show()
```

```python
a,b = np.unravel_index(score_mean_2.argmin(), score_mean_2.shape)
```

```python
print(f'The optimal number of N in KNN is {a+min_neighbor}, when number of lags is {b+n_lag_min}')
```
